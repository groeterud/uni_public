```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# read the audio file
sampling_freq, signal = wavfile.read('random_sound.wav')

#display the params
print('Signal shape:', signal.shape)
print('Signal Datatype:', signal.dtype)
print('Signal duration:', round(signal.shape[0] / float(sampling_freq), 2), 'seconds')

#normalize the signal
signal = signal / np.power(2, 15)

#extract the first 50 samples
signal = signal[:50]

#build the time axis in milliseconds
x_axis = 1000 * np.arange(0, len(signal), 1) / float(sampling_freq)

#plot the figure
plt.figure()
plt.plot(x_axis, signal, color='black')
plt.xlabel('Time (milliseconds)')
plt.ylabel('Amplitude')
plt.title('Input audio signal')
plt.show()

```
